```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

class LabelEncoderExt:
    def __init__(self):
        self.label_encoder = LabelEncoder()

    def fit(self, data):
        self.label_encoder = self.label_encoder.fit(list(data) + ['Unknown'])
        self.classes_ = self.label_encoder.classes_
        return self

    def transform(self, data):
        new_data = list(data)
        for unique_item in np.unique(data):
            if unique_item not in self.label_encoder.classes_:
                new_data = ['Unknown' if x == unique_item else x for x in new_data]
        return self.label_encoder.transform(new_data)
```

```python
le = LabelEncoderExt()

for col in X1.select_dtypes('object'):
  le.fit(X1[col])
  X1[col] = le.transform(X1[col])
```

*Klasifikasi Model*

```python
SEED = 42

# SUPPORT VECTOR CLASSIFICATION MODEL
from sklearn.svm import SVC
svc = SVC(random_state=SEED)

# LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

# GAUSSIAN NAIVE BAYES
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()


# LIGHTGB< CLASSIFIER
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier(random_state=SEED)
```

```python
from sklearn.model_selection import cross_val_score
models = {}
models['SVC'] = svc
models['Logistic Regression'] = lr
models['Naive Bayes Classifier'] = gnb
models['LightGBM Classifier'] = lgbm

scores = {}
for name in models:
  scores[name] = {}
  for scorer in ['precision', 'recall']:
    scores[name][scorer] = cross_val_score(models[name], X_latih_scaled, y_latih, cv=10, scoring=scorer)
```

```python
def line(name):
      return '*'*(25-len(name)//2)

for name in models:
  print(line(name), name, 'Model Validation', line(name))

  for scorer in ['precision','recall']:
    mean = round(np.mean(scores[name][scorer])*100,2)
    stdev = round(np.std(scores[name][scorer])*100,2)

    print('Mean {}:'.format(scorer),'\n', mean, '%','+-', stdev)
    print()
```

```python
for name in models:
    for scorer in ['precision','recall']:
        scores[name][scorer] = scores[name][scorer].mean()

scores = pd.DataFrame(scores).swapaxes('index','columns')*100
```
